# CRUD opérations suite...

Dans le chapitre 2, on a vu les opérations INSERT et SELECT.

Il nous reste deux autres opérations, UPDATE et DELETE.

Reprenons notre abse learning et le schéma public.

## UPDATE

On peut modifier des données d'une table avec l'opération UPDATE et des conditions.

**ATTENTION :** Sans conditions c'est l'ensemble des données de la table qui sera modifié.

Encore une fois, on peut accéder rapidement à la syntaxe d'une opération INSERT depuis pgadmin en faisant un clic droit sur une table > Scripts > UPDATE Script



On modifie la descrition du produit qui porte l'`id` 1.

`UPDATE public.product
	SET product_description = 'Le feu, source de chaleur'
	WHERE id = 1;`

On modifie le poids du produit qui porte l'`id` 2 et du crayon.

	`UPDATE public.product
	SET product_weight = 0.15
	WHERE id = 2 OR product_name = 'crayon';`

## DELETE

On peut supprimer des données d'une table avec l'opération DELETE et des conditions.

**ATTENTION :** Sans conditions c'est l'ensemble des données de la table qui sera supprimé.

Pour accéder à la syntaxe de la requête DELETE, clic droit sur une table > Scripts > DELETE Script.



On supprime la transaction qui porte l`id` 10.

`DELETE * FROM transaction WHERE id = 10;`

On supprime tous les produits.

`DELETE  FROM product`;

Cette requête renvoie une erreur car des produits sont dans la table `transaction`.

ERROR:  Key (id)=(3) is still referenced from table "transaction".update or delete on table "product" violates foreign key constraint "product_id_fkey" on table "transaction" 

ERROR:  update or delete on table "product" violates foreign key constraint "product_id_fkey" on table "transaction"
SQL state: 23503
Detail: Key (id)=(3) is still referenced from table "transaction".

Les contraintes de clés étrangères sécurise les tables `client` et `product`. Les client_id et product_id de la table `transaction` doivent appartenir aux tables `client` et `product` respectivement. On ne peut dont pas supprimer les données des tables `client` et `product` s'ils sont encore dans la table `transaction`.
