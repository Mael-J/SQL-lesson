# Modifier et supprimer des tables

Il est possible de modifier une table depuis l'interface graphique.

Clic droit sur la tabel > Properties

Dans l'onglet Columns, on peut modifier les noms de columns, leurs types, supprimer et ajouter des colonnes.



## Ajouter une colonne

On ajoute la colonne transaction_date  la table `transaction`, le type est date et la caleur par dfaut est aujourd'hui

`ALTER TABLE transaction
DROP transaction_date
ADD COLUMN transaction_date TIMESTAMP WITHOUT TIME ZONE NOT NULL DEFAULT current_date;`

## Supprimer une colonne

On supprime la colonne transaction_date.

`ALTER TABLE transaction
DROP transaction_date;`

## Supprimer une table

On peut supprimer une table depuis l'object explorer de pgAdmin.

clic droit sur la table > Delete



La commande SQL correspondante est

`DROP TABLE IF EXISTS transaction;`

## Supprimer une base de donnes

On peut supprimer une base de donnes depuis pdAdmin.

Clic droit sur la base > Delete



La commande SQL correspondante est :

`DROP DATABASE IF EXISTS learning;`
